# What are encodings

```python
before = "This is the euro symbol: €"

type(before)
```

```python
after = before.encode("utf-8", errors = "replace")

type(after)
```

```python
after
```

```python
print(after.decode("utf-8"))
```

```python
print(after.decode("ascii"))
```

```python
before = "This is the euro symbol: €"

after = before.encode("ascii", errors="replace")

print(after.decode("ascii"))

# print(after.decode("utf-8"))
```

```python
print(after.decode("utf-8"))
```
